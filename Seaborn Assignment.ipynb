```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.axes_style(x='fare',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.boxplot(x='class',y='age',data=titanic)
```

```python
sns.swarmplot(x='class',y='age',data=titanic)
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set1')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
sns.heatmap(titanic.corr())
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
sns.heatmap(titanic.corr(),cmap='rainbow')
```

```python
sns.heatmap(titanic.corr(),cmap='Set2')
```

```python
g=sns.FacetGrid(titanic,row='sex')
g.map(plt.hist,'age')
```

```python
g=sns.FacetGrid(titanic,col='sex')
g.map(plt.hist,'age')
```

```python
sns.displot(titanic['fare'],bins=30,color='pink')
```
